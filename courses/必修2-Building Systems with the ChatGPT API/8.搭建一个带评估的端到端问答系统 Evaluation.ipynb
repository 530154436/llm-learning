1. 检查输入，确认其是否能通过审核 API 的审核。

2. 如果通过了审核，我们将查找产品列表。

3. 如果找到了产品，我们将尝试查找它们的相关信息。

4. 我们使用模型回答用户提出的问题。

5. 我们将通过审核 API 对生成的答案进行审核。
